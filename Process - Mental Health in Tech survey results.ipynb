````python
#Load libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
#Read in cleaned csv
mh_survey = pd.read_csv("mh_survey.csv")
mh_survey
````

````python
#First want to look at family_history
history = mh_survey.groupby("family_history")["Respondent"].count()
history
````

````python
#Second look at values for treatment
treat = mh_survey.groupby("treatment")['Respondent'].count()
treat
````

````python
#Change values to 1's and 0's to be able to process correlation
mh_survey['family_history'] = mh_survey['family_history'].map({'Yes': 1, 'No': 0})
mh_survey['treatment'] = mh_survey['treatment'].map({'Yes': 1, 'No': 0})
mh_survey.head()
````

````python
#Now I want to see if there is a correlation between family_history and treatment
history_corr = mh_survey.corr()['family_history']['treatment']
print('Correlation between Family History and Sought Treatment is:{} '.format(history_corr))
````

```
There is a moderate positive correlation between having a family history of mental health issues to having sought treatment. I wanted to look at this especially as my step-dad most likely had a history of mental health disorders, but was born in 1945 when there wasn't much research, so he "didn't believe in it." However, my brothers both have mental health disorders now so it's interesting to see that correlation somewhat prove that mental health issues can be hereditary. 
```
